```
import pandas as pd
import plotly.graph_objects as go
```

```
#@title Load data

path = "/content/drive/My Drive/AI Index Visualization Project/DATASETS/GoogleTrends" #@param {type:"string"}
df_web_search_us = pd.read_csv(f"{path}/usa_web_search.csv")
df_news_search_us = pd.read_csv(f"{path}/usa_news_search.csv")
df_web_search_world = pd.read_csv(f"{path}/worldwide_web_search.csv")
df_news_search_world = pd.read_csv(f"{path}/worldwide_news_search.csv")

df_web_search_us['Month'] = pd.to_datetime(df_web_search_us['Month'])
df_web_search_us['Date'] = df_web_search_us.set_index('Month').index.year
df_web_search_us = df_web_search_us.drop('Month',axis=1).groupby('Date').agg('sum').reset_index()

df_news_search_us['Month'] = pd.to_datetime(df_news_search_us['Month'])
df_news_search_us['Date'] = df_news_search_us.set_index('Month').index.year
df_news_search_us = df_news_search_us.drop('Month',axis=1).groupby('Date').agg('sum').reset_index()

df_web_search_world['Month'] = pd.to_datetime(df_web_search_world['Month'])
df_web_search_world['Date'] = df_web_search_world.set_index('Month').index.year
df_web_search_world = df_web_search_world.drop('Month',axis=1).groupby('Date').agg('sum').reset_index()

df_news_search_world['Month'] = pd.to_datetime(df_news_search_world['Month'])
df_news_search_world['Date'] = df_news_search_world.set_index('Month').index.year
df_news_search_world = df_news_search_world.drop('Month',axis=1).groupby('Date').agg('sum').reset_index()

df_ai = pd.read_csv(f"{path}/ai_geoMap.csv")
df_ml = pd.read_csv(f"{path}/ml_geoMap.csv")
df_cs = pd.read_csv(f"{path}/cs_geoMap.csv")
df_nlp = pd.read_csv(f"{path}/nlp_geoMap.csv")
df_r = pd.read_csv(f"{path}/robotics_geoMap.csv")
```

```
#@title Web Search
AI = True #@param {type:"boolean"}
ML = True #@param {type:"boolean"}
NLP = True #@param {type:"boolean"}
Robotics = True #@param {type:"boolean"}
CS = True #@param {type:"boolean"}

Region = "Worldwide" #@param ["USA","Worldwide"]
if Region == "USA":
  df = df_web_search_us
else:
  df = df_web_search_world

fig = go.Figure()
colors = ['Blue','Red','Green','Black','Yellow']

if (AI):
  fig.add_trace(go.Scatter(x=df['Date'], y=df['Artificial intelligence'], name='Artificial intelligence', 
                         line_color=colors[0], hovertext=df['Artificial intelligence'], opacity=0.7))
if (ML):
  fig.add_trace(go.Scatter(x=df['Date'], y=df['Machine learning'], name='Machine learning', 
                         line_color=colors[1], hovertext=df['Machine learning'], opacity=0.7))

if (NLP):
  fig.add_trace(go.Scatter(x=df['Date'], y=df['Natural language processing'], name='Natural language processings', 
                         line_color=colors[2], hovertext=df['Natural language processing'], opacity=0.7))
if (Robotics):
  fig.add_trace(go.Scatter(x=df['Date'], y=df['Robotics'], name='Robotics', 
                         line_color=colors[3], hovertext=df['Robotics'], opacity=0.7))
if (CS):
  fig.add_trace(go.Scatter(x=df['Date'], y=df['Computer Science'], name='Computer Science', 
                         line_color=colors[4], hovertext=df['Computer Science'], opacity=0.7))



fig.add_layout_image(
    dict(
        source="https://pbs.twimg.com/profile_images/928845333349543937/xzRkcN08.jpg",
        xref="paper", yref="paper",
        x=-0.08, y=-0.28,
        sizex=0.23, sizey=0.23,
        xanchor="left", yanchor="bottom"
    )
)

fig.update_layout(
    title = {
        'text': 'Web Search',
        'y':0.9,
        'x':0,
        'xanchor': 'left',
        'yanchor': 'top'},
        titlefont= {
            "size": 30,
            
        },
    yaxis_title = "",
)
fig.show()
```

```
#@title ##### Country 
Field = "Machine Learning" #@param ["Artificial Intelligence","NLP","Machine Learning","Robotics","Computer Science"]
if Field == "Artificial Intelligence":
  df = df_ai.sort_values('Artificial Intelligence',ascending=False).head(10).sort_values('Artificial Intelligence')
  
elif Field == "NLP":
  df = df_nlp.sort_values('Natural language processing',ascending=False).head(10).sort_values('Natural language processing')
  
elif Field == "Machine Learning":
  df = df_ml.sort_values('Machine learning',ascending=False).head(10).sort_values('Machine learning')
  
elif Field == "Robotics":
  df = df_r.sort_values('Robotics',ascending=False).head(10).sort_values('Robotics')

elif Field == "Computer Science":
  df = df_cs.sort_values('Computer Science',ascending=False).head(10).sort_values('Computer Science')

col = list(df.columns)[1]
fig = go.Figure(data=[go.Bar(y=list(df['Country']), x=df[col],orientation='h')])

fig.add_layout_image(
    dict(
        source="https://pbs.twimg.com/profile_images/928845333349543937/xzRkcN08.jpg",
        xref="paper", yref="paper",
        x=-0.05, y=-0.26,
        sizex=0.23, sizey=0.23,
        xanchor="left", yanchor="bottom"
    )
)

fig.update_layout(
    title = {
        'text': 'Top Countries in {}'.format(Field),
        'y':0.9,
        'x':0,
        'xanchor': 'left',
        'yanchor': 'top'},
        titlefont= {
            "size": 30,
            
        },
    yaxis_title = "Country",
    xaxis_title = "Score"
)
fig.show()
```
