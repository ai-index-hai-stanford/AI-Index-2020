```
#@title Imports
import pandas as pd
import plotly.graph_objects as go
```

```
#@title Load data

path = "/content/drive/My Drive/AI Index Visualization Project/DATASETS/PubMed" #@param {type:"string"}
# load timeline data
df_ai = pd.read_csv(f"{path}/Artificial Intelligence/ai_timeline.csv")
df_ml = pd.read_csv(f"{path}/Machine Learning/ml_timeline.csv")
df_nlp = pd.read_csv(f"{path}/NLP/nlp_timeline.csv")
df_cv = pd.read_csv(f"{path}/Computer Vision/cv_timeline.csv")
df_r = pd.read_csv(f"{path}/Robotics/robotics_timeline.csv")

# load paper data
df_ai_paper = pd.read_csv(f"{path}/Artificial Intelligence/artificial_intelligence.csv")
df_ml_paper = pd.read_csv(f"{path}/Machine Learning/machine_learning.csv")
df_nlp_paper = pd.read_csv(f"{path}/NLP/nlp.csv")
df_cv_paper = pd.read_csv(f"{path}/Computer Vision/computer_vision.csv")
df_r_paper = pd.read_csv(f"{path}/Robotics/robotics.csv")

#pre
data_ai_paper = df_ai_paper[['First Author','Journal/Book','Publication Year','PMID']]
data_ml_paper = df_ml_paper[['First Author','Journal/Book','Publication Year','PMID']]
data_nlp_paper = df_nlp_paper[['First Author','Journal/Book','Publication Year','PMID']]
data_cv_paper = df_cv_paper[['First Author','Journal/Book','Publication Year','PMID']]
data_r_paper = df_r_paper[['First Author','Journal/Book','Publication Year','PMID']]
```

```
#@title #### Papers per year at PubMed
Field = "NLP" #@param ["Artificial intelligence","Machine Learning","NLP","Computer Vision","Robotics"]
if Field == "Artificial intelligence":
  df = df_ai
elif Field == "Machine Learning":
  df = df_ml
elif Field == "NLP":
  df = df_nlp
elif Field == "Computer Vision":
  df = df_cv
elif Field == "Robotics":
  df = df_r

fig = go.Figure([go.Bar(
    x=df['Year'], 
    y=df['Count'], 
    text=df['Count'], 
    textposition='auto')])

fig.add_layout_image(
    dict(
        source="https://pbs.twimg.com/profile_images/928845333349543937/xzRkcN08.jpg",
        xref="paper", yref="paper",
        x=-0.05, y=-0.26,
        sizex=0.23, sizey=0.23,
        xanchor="left", yanchor="bottom"
    )
)

fig.update_layout(
    title = f'Number of papers per year in {Field} at PubMed',  
    yaxis_title = 'Number of papers', 
    xaxis_tickmode = 'linear')
fig.show()
```

```
#@title ##### Top Authors
Field = "NLP" #@param ["Artificial intelligence","Machine Learning","NLP","Computer Vision","Robotics"]
if Field == "Artificial intelligence":
  df = df_ai_paper
elif Field == "Machine Learning":
  df = df_ml_paper
elif Field == "NLP":
  df = df_nlp_paper
elif Field == "Computer Vision":
  df = df_cv_paper
elif Field == "Robotics":
  df = df_r_paper
d = df.groupby(['First Author']).agg('count').reset_index().drop(['Journal/Book','Publication Year'],axis = 1).rename(columns={'PMID':'Number_of_papers'}).sort_values('Number_of_papers',ascending = False).head(10).sort_values('Number_of_papers')
fig = go.Figure([go.Bar(y=list(d['First Author']), x=d['Number_of_papers'],text=d['Number_of_papers'],orientation='h')])
fig.add_layout_image(
    dict(
        source="https://pbs.twimg.com/profile_images/928845333349543937/xzRkcN08.jpg",
        xref="paper", yref="paper",
        x=-0.05, y=-0.26,
        sizex=0.23, sizey=0.23,
        xanchor="left", yanchor="bottom"
    )
)

fig.update_layout(
    title = {
        'text': 'Top Authors at PubMed in {}'.format(Field),
        'y':0.9,
        'x':0,
        'xanchor': 'left',
        'yanchor': 'top'},
        titlefont= {
            "size": 26,
            
        },
    yaxis_title = "Authors",
    xaxis_title = "Number of Publications"
)
fig.show()
```

```
#@title ##### Top Journals/Books
Field = "NLP" #@param ["Artificial intelligence","Machine Learning","NLP","Computer Vision","Robotics"]
if Field == "Artificial intelligence":
  df = df_ai_paper
elif Field == "Machine Learning":
  df = df_ml_paper
elif Field == "NLP":
  df = df_nlp_paper
elif Field == "Computer Vision":
  df = df_cv_paper
elif Field == "Robotics":
  df = df_r_paper
d = df.groupby(['Journal/Book']).agg('count').reset_index().drop(['First Author','Publication Year'],axis = 1).rename(columns={'PMID':'Number_of_papers'}).sort_values('Number_of_papers',ascending = False).head(10).sort_values('Number_of_papers')
fig = go.Figure([go.Bar(y=list(d['Journal/Book']), x=d['Number_of_papers'],text=d['Number_of_papers'],orientation='h')])
fig.add_layout_image(
    dict(
        source="https://pbs.twimg.com/profile_images/928845333349543937/xzRkcN08.jpg",
        xref="paper", yref="paper",
        x=-0.05, y=-0.26,
        sizex=0.23, sizey=0.23,
        xanchor="left", yanchor="bottom"
    )
)

fig.update_layout(
    title = {
        'text': 'Top Journal/Book at PubMed in {}'.format(Field),
        'y':0.9,
        'x':0,
        'xanchor': 'left',
        'yanchor': 'top'},
        titlefont= {
            "size": 26,
            
        },
    yaxis_title = "Journal/Book",
    xaxis_title = "Number of Publications"
)
fig.show()
```
