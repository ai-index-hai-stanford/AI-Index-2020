```
import pandas as pd
import plotly.graph_objects as go
```

```
path = "/content/drive/My Drive/AI Index Visualization Project/DATASETS/Kaggle" #@param {type:"string"}

df_competition = pd.read_csv(f"{path}/Competitions.csv")
df_competition['EnabledDate'] = pd.to_datetime(df_competition['EnabledDate'])
df_competition = df_competition.set_index('EnabledDate')
df_competition.index = df_competition.index.year
df_competition = df_competition.reset_index()

df_forum = pd.read_csv(f"{path}/ForumTopics.csv")
df_forum['CreationDate'] = pd.to_datetime(df_forum['CreationDate'])
df_forum = df_forum.set_index('CreationDate')
df_forum.index = df_forum.index.year
df_forum = df_forum.reset_index()

df_tags = pd.read_csv(f"{path}/Tags.csv")
```

### Competitions

```
#@title #### Compititions per Category
df = df_competition[['EnabledDate','HostSegmentTitle','Id']]
df = df.groupby(['EnabledDate','HostSegmentTitle']).agg('count').rename(columns={'Id':'Number of competition'}).reset_index()
Year = "2018" #@param {type:"string"}
Year = int(Year)
d = df[df['EnabledDate']==Year].sort_values('Number of competition',ascending=False).head(20)
d = d.sort_values('Number of competition')
fig = go.Figure([go.Bar(y=d['HostSegmentTitle'], x=d['Number of competition'],orientation='h')])
fig.add_layout_image(
    dict(
        source="https://pbs.twimg.com/profile_images/928845333349543937/xzRkcN08.jpg",
        xref="paper", yref="paper",
        x=-0.07, y=-0.26,
        sizex=0.23, sizey=0.23,
        xanchor="left", yanchor="bottom"
    )
)

fig.update_layout(
    title = {
        'text': 'Competitions in {}'.format(Year),
        'y':0.9,
        'x':0,
        'xanchor': 'left',
        'yanchor': 'top'},
        titlefont= {
            "size": 30,
            
        },
    yaxis_title = "Competition type",
    xaxis_title = "Number of competitions"
)
fig.show()
```

```
#@title ##### Top competition based on number of submissions
Year = "2020" #@param {type:"string"}
Year = int(Year)
df = df_competition[df_competition['EnabledDate']==Year]
df = df[['Title','TotalSubmissions']].sort_values('TotalSubmissions',ascending=False).head(10).sort_values('TotalSubmissions')

fig = go.Figure(data=[go.Bar(y=list(df['Title']), x=df['TotalSubmissions'],orientation='h')])
fig.add_layout_image(
    dict(
        source="https://pbs.twimg.com/profile_images/928845333349543937/xzRkcN08.jpg",
        xref="paper", yref="paper",
        x=-0.05, y=-0.26,
        sizex=0.23, sizey=0.23,
        xanchor="left", yanchor="bottom"
    )
)

fig.update_layout(
    title = {
        'text': 'Top Competition in {}'.format(Year),
        'y':0.9,
        'x':0,
        'xanchor': 'left',
        'yanchor': 'top'},
        titlefont= {
            "size": 30,
            
        },
    yaxis_title = "Competitions",
    xaxis_title = "Number of submissions"
)
fig.show()
```

```
#@title Competitions over the years
df = df_competition[['EnabledDate','Title']].groupby('EnabledDate').agg('count').reset_index().rename(columns={'Title':'Number of competitions'})
df = df[df['EnabledDate']!=2001]
fig = go.Figure()
colors = ['Blue','Red','Green','Black','Yellow']
fig.add_trace(go.Scatter(x=df['EnabledDate'], y=df['Number of competitions'], name='Number of competitions', 
                         line_color=colors[0], hovertext=df['Number of competitions'], opacity=0.7))
fig.add_layout_image(
    dict(
        source="https://pbs.twimg.com/profile_images/928845333349543937/xzRkcN08.jpg",
        xref="paper", yref="paper",
        x=-0.06, y=-0.28,
        sizex=0.23, sizey=0.23,
        xanchor="left", yanchor="bottom"
    )
)

fig.update_layout(
    title = {
        'text': 'Competitions',
        'y':0.9,
        'x':0,
        'xanchor': 'left',
        'yanchor': 'top'},
        titlefont= {
            "size": 30,
            
        },
    yaxis_title = "Number of competitions",
)
fig.show()
```

### Forums

```
#@title ##### Top Forums  based on number of views
Year = "2019" #@param {type:"string"}
Year = int(Year)
df = df_forum[['CreationDate','Title','TotalViews']]
df[df['CreationDate']==Year]
df = df[['Title','TotalViews']].sort_values('TotalViews',ascending=False).head(10).sort_values('TotalViews')

fig = go.Figure(data=[go.Bar(y=list(df['Title']), x=df['TotalViews'],orientation='h')])
fig.add_layout_image(
    dict(
        source="https://pbs.twimg.com/profile_images/928845333349543937/xzRkcN08.jpg",
        xref="paper", yref="paper",
        x=-0.05, y=-0.26,
        sizex=0.23, sizey=0.23,
        xanchor="left", yanchor="bottom"
    )
)

fig.update_layout(
    title = {
        'text': 'Top Forums in {}'.format(Year),
        'y':0.9,
        'x':0,
        'xanchor': 'left',
        'yanchor': 'top'},
        titlefont= {
            "size": 30,
            
        },
    yaxis_title = "Forums",
    xaxis_title = "Number of Views"
)
fig.show()
```

```
#@title Forums over the years
df = df_forum[['CreationDate','Title']].groupby('CreationDate').agg('count').reset_index().rename(columns={'Title':'Number of forums'})

fig = go.Figure()
colors = ['Blue','Red','Green','Black','Yellow']
fig.add_trace(go.Scatter(x=df['CreationDate'], y=df['Number of forums'], name='Number of forums', 
                         line_color=colors[0], hovertext=df['Number of forums'], opacity=0.7))
fig.add_layout_image(
    dict(
        source="https://pbs.twimg.com/profile_images/928845333349543937/xzRkcN08.jpg",
        xref="paper", yref="paper",
        x=-0.06, y=-0.28,
        sizex=0.23, sizey=0.23,
        xanchor="left", yanchor="bottom"
    )
)

fig.update_layout(
    title = {
        'text': 'Forums',
        'y':0.9,
        'x':0,
        'xanchor': 'left',
        'yanchor': 'top'},
        titlefont= {
            "size": 30,
            
        },
    yaxis_title = "Number of forums",
)
fig.show()
```
