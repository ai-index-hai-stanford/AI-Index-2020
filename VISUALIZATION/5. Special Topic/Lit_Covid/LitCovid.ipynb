```
#@title Imports
import pandas as pd
import plotly.graph_objects as go
```

```
#@title Load Lit Covid Data
df = pd.read_csv("/content/drive/My Drive/AI Index Visualization Project/DATASETS/Lit_Covid/lit_covid.csv",sep='\t',error_bad_lines=False)
da = df.groupby('journal').agg('count')
da = da['pmid'].reset_index()
da = da.rename(columns={'pmid':'Number_of_papers'})
```

```
#@title Journals
da = da.sort_values('Number_of_papers',ascending=False).head(10).sort_values('Number_of_papers')
fig = go.Figure([go.Bar(y=list(da['journal']), x=da['Number_of_papers'],orientation='h')])
fig.add_layout_image(
    dict(
        source="https://pbs.twimg.com/profile_images/928845333349543937/xzRkcN08.jpg",
        xref="paper", yref="paper",
        x=-0.1, y=-0.26,
        sizex=0.23, sizey=0.23,
        xanchor="left", yanchor="bottom"
    )
)

fig.update_layout(
    title = {
        'text': 'Journals having most Covid Papers',
        'y':0.9,
        'x':0,
        'xanchor': 'left',
        'yanchor': 'top'},
        titlefont= {
            "size": 30,
            
        },
    yaxis_title = "Journals",
    xaxis_title = "Number of publications"
)
fig.show()
```
