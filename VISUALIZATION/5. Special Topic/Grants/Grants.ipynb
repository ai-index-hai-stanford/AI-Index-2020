```
#@title Imports
import pandas as pd
import plotly.graph_objects as go
```

```
#@title Load data
path = "/content/drive/My Drive/AI Index Visualization Project/DATASETS/Grants.gov" #@param {type:"string"}
df_ai = pd.read_csv(f"{path}/artificial intelligence.csv")
df_cv = pd.read_csv(f"{path}/computer_vision.csv")
df_ml = pd.read_csv(f"{path}/machine_learning.csv")
df_nlp = pd.read_csv(f"{path}/nlp.csv")
df_robotics = pd.read_csv(f"{path}/robotics.csv")
```

```
df_ai
```

```
#@title Top Opportunities

Field = "NLP" #@param ["Artificial Intelligence","Computer Vision","NLP","Machine Learning","Robotics"]
if Field == "Artificial Intelligence":
  data = df_ai
elif Field == "NLP":
  data = df_nlp
elif Field == "Machine Learning":
  data = df_ml
elif Field == "Robotics":
  data = df_robotics
elif Field == "Computer Vision":
  data = df_cv

data = data[['OPPORTUNITY NUMBER','ESTIMATED FUNDING']]
data = data.sort_values('ESTIMATED FUNDING',ascending=False).head(10).sort_values('ESTIMATED FUNDING')
fig = go.Figure(data=[go.Bar(name = "Total Funding", y=list(data['OPPORTUNITY NUMBER']), x=data['ESTIMATED FUNDING'],orientation='h')])
fig.add_layout_image(
    dict(
        source="https://pbs.twimg.com/profile_images/928845333349543937/xzRkcN08.jpg",
        xref="paper", yref="paper",
        x=-0.05, y=-0.26,
        sizex=0.23, sizey=0.23,
        xanchor="left", yanchor="bottom"
    )
)

fig.update_layout(
    title = {
        'text': 'Top Opportunities in {}'.format(Field),
        'y':0.9,
        'x':0,
        'xanchor': 'left',
        'yanchor': 'top'},
        titlefont= {
            "size": 20,
            
        },
    yaxis_title = "Opportunity",
    xaxis_title = "Total Fundings"
)
fig.show()
```

```
#@title Top Agencies based on number of opportunities

Field = "Artificial Intelligence" #@param ["Artificial Intelligence","Computer Vision","NLP","Machine Learning","Robotics"]
if Field == "Artificial Intelligence":
  data = df_ai
elif Field == "NLP":
  data = df_nlp
elif Field == "Machine Learning":
  data = df_ml
elif Field == "Robotics":
  data = df_robotics
elif Field == "Computer Vision":
  data = df_cv

data = data[['OPPORTUNITY TITLE','AGENCY NAME']]
data = data.groupby('AGENCY NAME').agg('count').reset_index().rename(columns={'OPPORTUNITY TITLE':'Opportunities'}).sort_values('Opportunities',ascending=False).head(10).sort_values('Opportunities')
fig = go.Figure(data=[go.Bar(name = "Number of opportunity", y=list(data['AGENCY NAME']), x=data['Opportunities'],orientation='h')])
fig.add_layout_image(
    dict(
        source="https://pbs.twimg.com/profile_images/928845333349543937/xzRkcN08.jpg",
        xref="paper", yref="paper",
        x=-0.05, y=-0.26,
        sizex=0.23, sizey=0.23,
        xanchor="left", yanchor="bottom"
    )
)

fig.update_layout(
    title = {
        'text': 'Top Agencies in {}'.format(Field),
        'y':0.9,
        'x':0,
        'xanchor': 'left',
        'yanchor': 'top'},
        titlefont= {
            "size": 30,
            
        },
    yaxis_title = "Agenties",
    xaxis_title = "Number of Opportunities"
)
fig.show()
```

```
#@title Top Agencies based on fundings

Field = "Artificial Intelligence" #@param ["Artificial Intelligence","Computer Vision","NLP","Machine Learning","Robotics"]
if Field == "Artificial Intelligence":
  data = df_ai
elif Field == "NLP":
  data = df_nlp
elif Field == "Machine Learning":
  data = df_ml
elif Field == "Robotics":
  data = df_robotics
elif Field == "Computer Vision":
  data = df_cv

data = data[['ESTIMATED FUNDING','AGENCY NAME']]
data = data.groupby('AGENCY NAME').agg('sum').reset_index().sort_values('ESTIMATED FUNDING',ascending=False).head(10).sort_values('ESTIMATED FUNDING')
fig = go.Figure(data=[go.Bar(name = "Total Funding", y=list(data['AGENCY NAME']), x=data['ESTIMATED FUNDING'],orientation='h')])
fig.add_layout_image(
    dict(
        source="https://pbs.twimg.com/profile_images/928845333349543937/xzRkcN08.jpg",
        xref="paper", yref="paper",
        x=-0.05, y=-0.26,
        sizex=0.23, sizey=0.23,
        xanchor="left", yanchor="bottom"
    )
)

fig.update_layout(
    title = {
        'text': 'Top Agencies in {} based on Fundings'.format(Field),
        'y':0.9,
        'x':0,
        'xanchor': 'left',
        'yanchor': 'top'},
        titlefont= {
            "size": 30,
            
        },
    yaxis_title = "Agenties",
    xaxis_title = "Total Funding"
)
fig.show()
```
