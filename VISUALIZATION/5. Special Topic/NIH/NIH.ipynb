```
import pandas as pd
import plotly.graph_objects as go
```

```
#@title Load data

path = "/content/drive/My Drive/AI Index Visualization Project/DATASETS/NIH" #@param {type:"string"}
df_ai_basic = pd.read_csv(f"{path}/Artificial Intelligence/Basic_data_AI.csv")
df_ai_adv = pd.read_csv(f"{path}/Artificial Intelligence/Adv_data_AI.csv")
df_ai_FY = pd.read_excel(f"{path}/Artificial Intelligence/ai_FY.xlsx")
df_ai_country = pd.read_excel(f"{path}/Artificial Intelligence/ai_country.xlsx")
df_ai_state = pd.read_excel(f"{path}/Artificial Intelligence/state.xlsx")

df_ml_adv = pd.read_csv(f"{path}/Machine Learning/Adv_data_machine_learning.csv")
df_ml_basic = pd.read_csv(f"{path}/Machine Learning/Basic_data_Machine_Learning.csv")
df_ml_FY = pd.read_excel(f"{path}/Machine Learning/ml_FY.xlsx")
df_ml_country = pd.read_excel(f"{path}/Machine Learning/ml_country.xlsx")
df_ml_state = pd.read_excel(f"{path}/Machine Learning/ml_state.xlsx")

df_nlp_adv = pd.read_csv(f"{path}/NLP/Adv_data_nlp.csv")
df_nlp_basic = pd.read_csv(f"{path}/NLP/basic_data_nlp.csv")
df_nlp_FY = pd.read_excel(f"{path}/NLP/nlp_FY.xlsx")
df_nlp_country = pd.read_excel(f"{path}/NLP/nlp_country.xlsx")
df_nlp_state = pd.read_excel(f"{path}/NLP/nlp_state.xlsx")

df_robotics_adv = pd.read_csv(f"{path}/Robotics/Adv_data_robotics.csv")
df_robotics_basic = pd.read_csv(f"{path}/Robotics/basic_data_robotics.csv")
df_robotics_FY = pd.read_excel(f"{path}/Robotics/robotics_FY.xlsx")
df_robotics_country = pd.read_excel(f"{path}/Robotics/country.xlsx")
df_robotics_state = pd.read_excel(f"{path}/Robotics/state.xlsx")
```

* Artificial intelligence:
 - FY : 1985 to 2020

* Robotics
 - FY = 2017 to 2020

```
#@title Administering Institute/Center per year Fundings

Year = "2020" #@param {type:"string"}
Year = int(Year)
Field = "Artificial Intelligence" #@param ["Artificial Intelligence","NLP","Machine Learning","Robotics"]
if Field == "Artificial Intelligence":
  data_fund = df_ai_basic
elif Field == "NLP":
  data_fund = df_nlp_basic
elif Field == "Machine Learning":
  data_fund = df_ml_basic
elif Field == "Robotics":
  data_fund = df_robotics_basic
      
data_fund = data_fund[['Administering  IC','FY', 'FY Total Cost','FY Total Cost  (Sub Projects)']]
data_fund.columns = data_fund.columns.str.replace(' ', '_')
data_fund = data_fund.groupby(['Administering__IC','FY']).agg('sum').reset_index()
data_fund = data_fund[data_fund['FY']==Year]
data_fund = data_fund.sort_values('FY_Total_Cost',ascending=False).head(10).sort_values('FY_Total_Cost')

fig = go.Figure(data=[go.Bar(name = "Total Funding", y=list(data_fund['Administering__IC']), x=data_fund['FY_Total_Cost'],orientation='h'),
                      go.Bar(name = "Sub Project Funding",y=list(data_fund['Administering__IC']), x=data_fund['FY_Total_Cost__(Sub_Projects)'],orientation='h')
                      ])
fig.add_layout_image(
    dict(
        source="https://pbs.twimg.com/profile_images/928845333349543937/xzRkcN08.jpg",
        xref="paper", yref="paper",
        x=-0.05, y=-0.26,
        sizex=0.23, sizey=0.23,
        xanchor="left", yanchor="bottom"
    )
)

fig.update_layout(
    title = {
        'text': 'Administering Institute/Center in {} in {}'.format(Field,Year),
        'y':0.9,
        'x':0,
        'xanchor': 'left',
        'yanchor': 'top'},
        titlefont= {
            "size": 30,
            
        },
    yaxis_title = "Institute / Center",
    xaxis_title = "Funding"
)
fig.show()
```
