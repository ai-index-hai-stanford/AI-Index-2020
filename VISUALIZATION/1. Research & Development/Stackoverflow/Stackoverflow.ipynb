# Frameworks

```
#@title ##### Comparing questions on different frameworks
Framework = "tensorflow" #@param ["tensorflow","keras","pytorch"]
Questions = "unanswered" #@param ["unanswered","popular"]
Title = "Top unanswered questions on tensorflow" #@param {type : "string"}
df = pd.read_csv(f"{path}/Frameworks/{Framework.capitalize()}/{Framework}_{Questions}.csv")
da = df[['view_count','title']].sort_values('view_count',ascending=False).head(10)
da = da.sort_values('view_count')
fig = go.Figure([go.Bar(y=da['title'], x=da['view_count'],orientation='h')])
fig.add_layout_image(
    dict(
        source="https://pbs.twimg.com/profile_images/928845333349543937/xzRkcN08.jpg",
        xref="paper", yref="paper",
        x=-0.2, y=-0.26,
        sizex=0.23, sizey=0.23,
        xanchor="left", yanchor="bottom"
    )
)
if Questions == "unanswered":   
  xtitle = "Views"
else:
  xtitle = "Score"
fig.update_layout(
    title = {
        'text': Title,
        'y':0.9,
        'x':0,
        'xanchor': 'left',
        'yanchor': 'top'},
        titlefont= {
            "size": 30,
            
        },
    yaxis_title = "Questions",
    xaxis_title = xtitle,
)
fig.show()
```

```
#@title Frameworks comparison line chart
import plotly.graph_objects as go
df_framework = pd.read_csv(f"{path}/Frameworks/frameworks_questions_count.csv")
df_framework = df_framework.drop('Unnamed: 0',axis = 1).set_index('Date').sort_index()
c = list(df_framework.columns)
fig = go.Figure()
colors = ['Blue','Red','Green','Black']
i=0
for col in c:
  fig.add_trace(go.Scatter(x=df_framework.index, y=df_framework[col], name=f'{col}', 
                         line_color=colors[i], hovertext=df_framework[col], opacity=0.7))
  i+=1
  i = i%len(colors)

fig.add_layout_image(
    dict(
        source="https://pbs.twimg.com/profile_images/928845333349543937/xzRkcN08.jpg",
        xref="paper", yref="paper",
        x=-0.08, y=-0.28,
        sizex=0.23, sizey=0.23,
        xanchor="left", yanchor="bottom"
    )
)

fig.update_layout(
    title = {
        'text': 'Framework Comparison',
        'y':0.9,
        'x':0,
        'xanchor': 'left',
        'yanchor': 'top'},
        titlefont= {
            "size": 30,
            
        },
    yaxis_title = "Number of questions",
)
fig.show()
```

```
#@title Growth chart
df_framework = pd.read_csv(f"{path}/Frameworks/frameworks_questions_count.csv")
df_framework = df_framework.drop('Unnamed: 0',axis = 1).set_index('Date').sort_index()
Framework = "keras" #@param ["tensorflow","keras","pytorch"]
if Framework == "pytorch":
  df = df_framework[['Pytorch_Question_Count']]
  df = df[df['Pytorch_Question_Count']!=0]
  col = list(df.describe().columns)
  df = df.reset_index()
  index = int(df[df['Date']==2017][col[0]])
elif Framework == "keras":
  df = df_framework[['Keras_Question_Count']]
  df = df[df['Keras_Question_Count']!=0]
  col = list(df.describe().columns)
  df = df.reset_index()
  index = int(df[df['Date']==2013][col[0]])
else:
  df = df_framework[['Tensorflow_Question_Count']]
  df = df[df['Tensorflow_Question_Count']!=0]
  col = list(df.describe().columns)
  df = df.reset_index()
  index = int(df[df['Date']==2011][col[0]])

df = df.set_index('Date')
time = list(df.index)
a = {time[0]:1}
for i in range(1,len(time)):
  a[time[i]] = (((int(df[df.index==time[i]][col[0]])/ index) - 1) * 1)+1
a = pd.DataFrame(a.items(), columns=['Date', 'number_of_ques'])
fig = go.Figure()
button_layer_1_height = 1.08
fig.add_layout_image(
    dict(
        source="https://pbs.twimg.com/profile_images/928845333349543937/xzRkcN08.jpg",
        xref="paper", yref="paper",
        x=-0.08, y=-0.36,
        sizex=0.25, sizey=0.25,
        xanchor="left", yanchor="bottom"
    )
)
fig.add_trace(go.Scatter(x=a['Date'], y=a['number_of_ques'], 
                        hovertext=a['number_of_ques']))

fig.update_layout(
    title = {
        'text':f'Growth in {Framework}',
        'y':0.9,
        'x':0,
        'xanchor': 'left',
        'yanchor': 'top'
        },
        titlefont = {
            "size": 30,    
        },
        xaxis_rangeslider_visible=True, 
 
        yaxis_title='Number of questions')
fig.show()
```
