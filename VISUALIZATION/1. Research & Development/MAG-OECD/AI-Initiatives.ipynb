```
#@title AI Initiatives
df = pd.read_csv(f'{data_path}/AI_Initiatives.csv')
fig = px.choropleth(df, locations="Country_Code",
                    color="AI_Initiatives", 
                    hover_name="Country", 
                    color_continuous_scale=px.colors.sequential.Viridis)
fig.update_layout(
    title = {
        'text': 'AI Initiatives',
        'y':0.9,
        'x':0,
        'xanchor': 'left',
        'yanchor': 'top'},
        titlefont= {
            "size": 30,
            
        },
    width=2000,
    height=900,
)
fig.show()
```

```
#@title Top Countries with most AI Initiatives

df = pd.read_csv(f'{data_path}/AI_Initiatives.csv')
Number_of_countries = 15 #@param {type:"slider", min:1, max:25, step:1}

da = df.sort_values('AI_Initiatives',ascending=False).head(Number_of_countries).sort_values('AI_Initiatives')
fig = go.Figure(data=[go.Bar(y=list(da['Country']), x=da['AI_Initiatives'],orientation='h')])

fig.add_layout_image(
    dict(
        source="https://pbs.twimg.com/profile_images/928845333349543937/xzRkcN08.jpg",
        xref="paper", yref="paper",
        x=-0.05, y=-0.26,
        sizex=0.23, sizey=0.23,
        xanchor="left", yanchor="bottom"
    )
)

fig.update_layout(
    title = {
        'text': 'Top countries having most AI Initiatives',
        'y':0.9,
        'x':0,
        'xanchor': 'left',
        'yanchor': 'top'},
        titlefont= {
            "size": 30,
            
        },
    yaxis_title = "Country",
    xaxis_title = "AI Initiatives"
)
fig.show()
```

#### Country Analysis

```
# @title Select Country
Country = "United States" #@param ['Argentina', 'Australia', 'Austria', 'Germany', 'China', 'Greece', 'Belgium', 'Brazil', 'Bulgaria', 'Canada', 'Chile', 'Colombia', 'Costa Rica', 'Croatia', 'Cyprus', 'Czech Republic', 'Denmark', 'Egypt', 'Estonia', 'Finland', 'France', 'Hungary', 'Iceland', 'India', 'Ireland', 'Israel', 'Italy', 'Japan', 'Kazakhstan', 'Korea', 'Latvia', 'Lithuania', 'Luxembourg', 'Malta', 'Mexico', 'Morocco', 'New Zealand', 'Netherlands', 'Norway', 'Poland', 'Portugal', 'Romania', 'Russian Federation', 'Saudi Arabia', 'Singapore', 'Serbia', 'Slovak Republic', 'Slovenia', 'South Africa', 'Spain', 'Sweden', 'Switzerland', 'Thailand', 'Turkey', 'United States', 'United Kingdom', 'Uruguay', 'Vietnam', 'United Arab Emirates', 'European Union']
c = country[Country]
df_country = pd.read_csv(f'{data_path}/{c}-policy-initiatives.csv')
```

```
#@title Top Themes
da = df_country[['Policy initiative ID','Theme(s)']].groupby('Theme(s)').agg('count').reset_index().rename(columns={'Policy initiative ID':'AI_Initiatives'}).sort_values('AI_Initiatives',ascending=False).head(10).sort_values('AI_Initiatives')
fig = go.Figure(data=[go.Bar(y=list(da['Theme(s)']), x=da['AI_Initiatives'],orientation='h')])

fig.add_layout_image(
    dict(
        source="https://pbs.twimg.com/profile_images/928845333349543937/xzRkcN08.jpg",
        xref="paper", yref="paper",
        x=-0.05, y=-0.26,
        sizex=0.23, sizey=0.23,
        xanchor="left", yanchor="bottom"
    )
)

fig.update_layout(
    title = {
        'text': f'Top Theme(s) for AI Initiatives in {Country}',
        'y':0.9,
        'x':0,
        'xanchor': 'left',
        'yanchor': 'top'},
        titlefont= {
            "size": 30,
            
        },
    yaxis_title = "Theme(s)",
    xaxis_title = "AI Initiatives"
)
fig.show()
```
