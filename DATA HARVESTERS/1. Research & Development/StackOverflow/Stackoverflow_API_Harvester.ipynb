```
#@title ##### Unanswered/ Popular Questions
import json
import requests
from pprint import pprint
import csv
import pandas as pd
from datetime import datetime

file_name = "MNIST_unanswered" #@param {type:"string"}
tags = "pytorch" #@param {type:"string"}
questions = "popular" #@param ["unanswered","popular"]
if questions == "unanswered":
  url = f"https://api.stackexchange.com/2.2/questions/unanswered?order=desc&sort=votes&tagged={tags}&site=stackoverflow&filter=!)rh-4yRhEc9jbDIpE91u"
else:
  url = f"https://api.stackexchange.com/2.2/questions?order=desc&sort=votes&tagged={tags}&site=stackoverflow&filter=!)rh-4yRhEc9jbDIpE91u"
response  = requests.get(url)
query = response.json()
json_object = json.dumps(query, indent = 6) 
with open('stack.json', "w") as outfile:
  outfile.write(json_object) 
with open('stack.json') as json_file: 
  data = json.load(json_file) 
employee_data = data['items'] 
data_file = open('data_file.csv', 'w') 
csv_writer = csv.writer(data_file) 
count = 0

for emp in employee_data: 
  if count == 0: 

    # Writing headers of CSV file 
    header = emp.keys() 
    csv_writer.writerow(header) 
    count += 1

  # Writing data of CSV file 
  csv_writer.writerow(emp.values()) 

data_file.close() 
df = pd.read_csv('data_file.csv', error_bad_lines=False)
df = df[['title','score','view_count','creation_date','answer_count']]
df = df.dropna()
l = list(df['creation_date'])
for i in range(len(l)):
  l[i] = int(l[i])
  l[i] = datetime.utcfromtimestamp(l[i]).strftime('%Y-%m-%d')
df['Date'] = l
df = df.drop('creation_date',axis=1)
df = df[['title','score','view_count']]
df.to_csv(f'/{file_name}.csv')
```
