```
import requests
from bs4 import BeautifulSoup
import pandas as pd

Page_start = "1" #@param{type:"string"}
Page_end = "19" #@param{type:"string"}
tags = "sklearn-pandas" #@param{type:"string"}
pages = list(range(int(Page_start),int(Page_end)+1))
questions = []
votes = []
views = []
date = []

for i in pages:
    print("Page-",i," Scrapped")
    res = requests.get("https://stackoverflow.com/questions/tagged/{}?tab=newest&page=".format(tags)+str(i)+"&pagesize=50")
    soup = BeautifulSoup(res.text,"html.parser")
    question = soup.select(".question-summary")
    for que in question:
        
        try:
            if que.select_one(".relativetime").has_attr('title') and que.select_one(".views").has_attr('title'):
                q = que.select_one(".question-hyperlink").getText()
                vote_count = que.select_one(".vote-count-post").getText()
                view = que.select_one(".views").attrs['title'].split(' ')[0]
                d = que.select_one(".relativetime").attrs['title'].split(' ')[0]
                questions.append(q)
                votes.append(vote_count)
                views.append(view)
                date.append(d)
        except:
            pass

data = pd.DataFrame({"Question":questions,"Votes":votes,"Views":views,"Date":date})
data.to_csv(f"{tags}.csv",encoding='utf-8',index = False)
# data.to_csv("{}.csv".format(tags), encoding='utf-8',index = False)
```
